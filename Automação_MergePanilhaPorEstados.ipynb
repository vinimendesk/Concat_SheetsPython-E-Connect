```python
# Importando bibliotecas.
import pandas as pd
import openpyxl
```

```python
path_empresasAbertas = "Planilhas/Cópia de EMPRESAS ABERTAS 21-02.xlsx"
# Abrindo planilha principal "Empresas Abertas".
empresas_abertas = pd.read_excel(path_empresasAbertas, sheet_name=None)

empresas_abertas
```

```python
# Abrindo planilhas gerais para mesclar com a principal "Empresas Abertas".
planilha9 = pd.read_excel("Planilhas/21-02-2025 9h.xlsx")
planilha10 = pd.read_excel("Planilhas/21-02-2025 10h.xlsx")
planilha12 = pd.read_excel("Planilhas/21-02-2025 12h.xlsx")
planilha13 = pd.read_excel("Planilhas/21-02-2025 13h.xlsx")
planilha14 = pd.read_excel("Planilhas/21-02-2025 14h.xlsx")

# Lista contendo todas as planilhas.
todasPlanilhas = [planilha9, planilha10, planilha12, planilha13, planilha14]

df_todasPlanilhas = pd.concat(todasPlanilhas, ignore_index=True)

df_todasPlanilhas
```

```python
# Armazena todos os estados presentes nas planilhas.
todosEstados = df_todasPlanilhas["UF"].unique() 

# Abre a planilha no modo de editor excel usando a biblioteca openpyxl.
with pd.ExcelWriter("Planilhas/EMPRESAS ABERTAS 21-02.xlsx", engine='openpyxl') as writer:  
    # Percorre todos os dados(df) de todas as abas (estado).
    for estado in todosEstados:
        # Filtrar tabelas pelo estado.
        dados_estados = df_todasPlanilhas[df_todasPlanilhas["UF"] == estado]
        # Filtrando somente colunas necessárias.
        df_estado = dados_estados[["UF", "Municipio", "CNPJ", "Razao Social", "Telefones", "E-mail"]]
        # Verifica se o estado já existe na planilha principal.
        if estado in empresas_abertas:
            # Faz a adição de dados a planilha principal.
            df_atualizado = pd.concat([empresas_abertas[estado], dados_estados], ignore_index=True)
        else: 
            df_atualizado = df_estado
        # Salva o arquivo.
        df_atualizado.to_excel(writer, sheet_name=estado, index=False)
            
```
